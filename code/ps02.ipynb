```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy
import math

%precision 3
%matplotlib inline
```

The shape of bartlett and hamming looks pretty similar except the bartlett's amplitude peak is at 160 while the hamming peak is at around 130. Hamming also seems to have more leakage toward the lower amplitude.

###Question 2
Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly. Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
class SawtoothChirp(thinkdsp.Chirp):
    
    def _evaluate(self, ts, freq):
        # bring phase from the chirp
        dts = numpy.diff(ts)
        dps = (math.pi*2) * freq * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        
        # replace cycle with phases calculated from above
        cycles = phases/(math.pi*2)
        frac, _ = numpy.modf(cycles) #just from sawtooth
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys 
```

```python
signal = SawtoothChirp(start = 440, end = 880)
wave = signal.make_wave(duration =1)
segment = wave.segment(0,.01)
segment.plot()
wave.make_audio()
```

SawtoothChirp was created by combining evaluate function for both Sawtooth and Chirp.
The sound is definitely interesting and sounds like the Star Trek emergency noise.

```python
spectogram = wave.make_spectrogram(seg_length = 512)
spectogram.plot()
```

The dominant band seems to go from 440Hz to 880Hz. The aliasing effect can be seen by observing the faint bands above the dominant band.

###Question 3
In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.
Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?
Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
class TromChirp(thinkdsp.Chirp):
    def evaluate(self, ts):
        
        start, end = 1.0/(self.start), 1.0/(self.end)
        begin = numpy.linspace(start, end, len(ts)/2)
        last = numpy.linspace(end, start, len(ts)/2)
        freqs = numpy.hstack((begin, last))
        return self._evaluate(ts, 1/freqs)

signal = TromChirp(start = 262, end = 349)    
wave = signal.make_wave()
wave.make_audio()
```

We created TromChirp from Chirp but implementing tricks in ExpoChirp. The frequency varies inversely compared to time so the start and end is inverse of self.start and self.end. We also invert the frequency at the end to correctly evaluate the wave. TromChirp divides ts by half and make the first half go from start to end and the second half go from end to start. Then we use hstack to append these two halves.
The sound produced definitely sounds like a trombone glissando from C3 to F3 and back down to C3

###Question 6
In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
rainfall_spec = rainfall_seg.make_spectrum()
rainfall_spec.plot_power(linewidth=1,color = 'blue')

thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 ylim=[1e0, 1e15],
                 legend=False)
```

```python
def bartletts_method(wave, K):
    
    # k = N/M
    segment_size = numpy.linspace(0, len(wave.ys), K+1) # divide into K segments
    parts = numpy.zeros((K,len(wave.ys)/K)) # create a numpy array 
    waves = []
    segment_spectrum = []
    sumSpect = []
    for i in range(K):
        parts[i] =  wave.ys[segment_size[i]:segment_size[i+1]]
        waves.append(thinkdsp.Wave(parts[i],11025))
        segment_spectrum.append(waves[i].make_spectrum())
        sumSpect.append(abs(segment_spectrum[i].hs**2)/K)  # square the spectrum value  and divide it by K
        
    average = sum(sumSpect)/len(sumSpect) # find the average
    return thinkdsp.Spectrum(average, 44100)  #create a new spectrum from the average value

rainfall_seg1 = rainfall_seg
spectrum = bartletts_method(rainfall_seg1, 100)
spectrum.plot(high= 50)
        

rainfall_seg.ys = numpy.bartlett(len(rainfall_seg.ys))
rainfall_spec = rainfall_seg.make_spectrum()
rainfall_spec.plot_power(linewidth=1,color = 'red')

thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 ylim=[1e-7, 1e12],
                 legend=False)
```

We plot both Bartletts window that numpy provides and the Bartletts method. It seems that barlett method cleans up the readings from the bartlett windown which makes sense because in Bartletts, we section the wave and find the average of the spectrum. It seems that power to frequency relationship stays constant then at 10Hz, power denisty seems to drop.
